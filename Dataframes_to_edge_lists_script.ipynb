```python
#create a function that will identify the indexes to which the authors are null in a dataframe
def null_indexes(df):
    nulls = df.loc[pd.isnull(df.Authors),]
    nulls = nulls.index
    return nulls
```

```python
#function that will transform dataframes to csv files
def dataframe_to_csv(df):
    #create a null dataframe
    edge_list = pd.DataFrame(columns=['From','To'])
    for j in range(0,len(df)):
        #if the index shows that there are no authors, it skips the iteration
        if j in null_indexes(df):
            continue
        else:
            #find all combinations of 2 authors per observation
            temp = list(combinations(df.iloc[j,1], 2))
            #insert the combinations in a temporary dataframe
            temp_df = pd.DataFrame(temp, columns=['From','To'])
            #append the temporary dataframe to the edge_list df
            edge_list = edge_list.append(temp_df, ignore_index = True)
    
    #find the weights of each combination by calculating their frequencies
    weights = edge_list.groupby(['From', 'To']).size().reset_index()
    #rename the column as weight
    weights.rename(columns={0: 'Weight'}, inplace=True)
    #export the dataframe to csv format without their indexes
    weights.to_csv('edge_list_'+str(int(df.Year.unique()))+'.csv', index=False)
```

```python
#transformation of all datasets (one for each year) to csv format
dataframe_to_csv(authors_2016)
dataframe_to_csv(authors_2017)
dataframe_to_csv(authors_2018)
dataframe_to_csv(authors_2019)
dataframe_to_csv(authors_2020)
```
